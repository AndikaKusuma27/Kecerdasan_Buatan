```python
import keras
import numpy as np
from sklearn.datasets import load_iris
dataset=load_iris()
print(dataset)
```

```python
#langkah1 mengetahui data itu dari mana
print(type(dataset))
print(len(dataset))
#menyisir format dataset tersebut. dengan objek.
print(dataset.keys())
#mencoba untuk mencari nilai yang ada di riwayat dengan atribut
print(dataset['data'][0:5])
print(dataset['target'][0:5])
print(dataset['target_names'][0:5])
print(dataset['DESCR'])
print(dataset['feature_names'])
print(dataset['filename'])
#lanjutkan untuk menentukan x dan y; y memiliki ketergantungan dari x
x=dataset['data']
print(len(x)) #untuk memeriksa panjangnya dataset
y=dataset['target']
print(len(y)) #untuk memeriksa panjangnya targets
print(x[0])
print(y[0])
```

```python
#langkah 2 Ubah Y menjadi satu vektor yang dikodekan teratas
#kode teratas adalah penggunaan karena y termasuk label (0,1,2)- yang mana kategori data dengan bukan hubungan ordinal
from keras.utils import to_categorical
Ny=len(np.unique(y))
print(Ny)
Y=to_categorical(y,num_classes=Ny)
print(Y[0:5])
```

```python
#langkah 3 : sekarang kami akan memecah data sampai 2 bagian - 1 untuk data latih dan lainnya untuk untuk data testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,Y,test_size=0.10,shuffle=True)
print(x_train[0:5])
print(x_test[0:5])
print(y_train[0:5])
print(y_test[0:5])
```

```python
#langkah 4 :kemudian menormlakan data
from sklearn.preprocessing import StandardScaler
#selesai menormalkan data lalu membedakan antara tertinggi dan terendah suatu data point yang tidak besar juga
import numpy as np
scaler=StandardScaler()
#untuk mencari rata-rata dan standar dev
scaler.fit(x_train)

#mengubah data menjadi bentuk dimana rata-rata dari data adalah 0 dan std dev adalah 1
X_train=scaler.transform(x_train)
X_test=scaler.transform(x_test)
print(np.amax(X_train,axis=0))
print(np.amin(X_train,axis=0))
```

```python
#langkah ke5 : terakhir membuat model dari keras
!pip install tensorflow
import keras
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(20, input_dim=X_train.shape[1], activation='relu'))
#Dropout untuk menghindari overfitting
keras.layers.Dropout(0.2)
model.add(Dense(20, activation='relu'))
keras.layers.Dropout(0.2)
model.add(Dense(20, activation='relu'))
keras.layers.Dropout(0.2)
#untuk mengkalsifikasikan masalah, biasanya menggunakan softmax untuk aktifasi fungsi di akhir lapisan
model.add(Dense(3, activation='softmax'))
#kompilasi model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#langkah ke 6 : mengetahui model
model.summary()
```

```python
#langkah ke 7 : melatih model
#Batch ukuran,epoch dan validasi membelah semua hyper parameters
model.fit(X_train, y_train, epochs=195, batch_size=80, validation_split=0.1)
```

```python
#langkah 8 : mengambil akurasi dari model di data uji
testing=model.evaluate(X_test, y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1]+' dari model data uji', testing[1]*100))
```

```python
#langkah 9 : mengevaluasi model lain
from sklearn.metrics import classification_report, confusion_matrix
predictions = np.argmax(model.predict(X_test), axis=1)
Y_test = np.argmax(y_test, axis=1)
# mengambil confusion matrix
print(confusion_matrix(Y_test, predictions))
# mengambil nilai dari semua evaluation matrics
print(classification_report(Y_test,predictions))
```
